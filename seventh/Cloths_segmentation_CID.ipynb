If you want to use cloth segmentation on webcam captured image, run below code

```python
import cv2
import os
from PIL import Image
from IPython.display import display

def take_photo(filename='photo.jpg'):
    cam = cv2.VideoCapture(0)  # Open the default webcam (0)
    
    if not cam.isOpened():
        print("Error: Could not access the webcam.")
        return None

    print("Press 'SPACE' to capture the photo, 'ESC' to exit.")

    while True:
        ret, frame = cam.read()
        if not ret:
            print("Failed to capture image")
            break
        
        cv2.imshow("Webcam", frame)

        key = cv2.waitKey(1) & 0xFF
        if key == 32:  # Press SPACE to capture
            save_directory = os.path.join(os.getcwd(), "captured_images")
            os.makedirs(save_directory, exist_ok=True)  # Ensure directory exists

            save_path = os.path.join(save_directory, filename)
            cv2.imwrite(save_path, frame)
            print(f"Photo saved to: {save_path}")
            break
        elif key == 27:  # Press ESC to exit without capturing
            print("Photo capture cancelled.")
            break

    cam.release()
    cv2.destroyAllWindows()
    return save_path

# Run the function and display the captured image
try:
    img_path = take_photo("captured_photo.jpg")
    if img_path:
        print(f"Saved to {img_path}")

        # Show the captured image using PIL
        img = Image.open(img_path)
        display(img)
except Exception as err:
    print(f"Error: {str(err)}")
```

```python
from iglovikov_helper_functions.utils.image_utils import load_rgb, pad, unpad
from iglovikov_helper_functions.dl.pytorch.utils import tensor_from_rgb_image
```

```python
from cloths_segmentation.pre_trained_models import create_model
```

```python
model = create_model("Unet_2020-10-30")
```

```python
model.eval();
```

```python
image = load_rgb('F:\clothfoldingmachine\seventh\image.jpeg')
```

```python
#print(img)
print(img)

#files.view(keys[-1])
#image = np.array(image)
```

```python
#trying to 
#imshow(image)
#files.download('pankaj.png')
imshow(image)
```

```python
transform = albu.Compose([albu.Normalize(p=1)], p=1)
```

```python
padded_image, pads = pad(image, factor=32, border=cv2.BORDER_CONSTANT)
```

```python
x = transform(image=padded_image)["image"]
x = torch.unsqueeze(tensor_from_rgb_image(x), 0)
```

```python
with torch.no_grad():
  prediction = model(x)[0][0]
```

```python
mask = (prediction > 0).cpu().numpy().astype(np.uint8)
```

```python
mask = unpad(mask, pads)
```

```python
imshow(mask)
```

```python
dst = cv2.addWeighted(image, 1, (cv2.cvtColor(mask, cv2.COLOR_GRAY2RGB) * (0, 255, 0)).astype(np.uint8), 0.5, 0)
```

```python
imshow(dst)
```

```python
imshow(np.hstack([image, cv2.cvtColor(mask, cv2.COLOR_GRAY2RGB) * 255, dst]))
```
